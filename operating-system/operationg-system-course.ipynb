```python
empty_pixels = np.zeros(shape=(600, 600))
```

```python
drawed = []
```

```python
def draw_pixel(x, y):
    print(f'draw({x},{y})')
    
    global drawed, empty_pixels
    
    drawed.append((x, y))
    
    for _x, _y in drawed:
        empty_pixels[x, y] = 1

def draw_line(x1, y1, x2, y2):

    a, b = 0, 0
    diff = 0
    dx, dy = x2 - x1, y2 - y1
    
    #global drawed
    #drawed = []
    
    while a <= dx and b <= dy:
        
        draw_pixel(x1+a, y1+b)
        
        if diff <= 0: 
            a += 1
            diff += dy
        else:
            b += 1
            diff -= dx
```

```python
draw_line(*(1, 0), *(20, 40))
```

```python
plt.imshow(empty_pixels, cmap="Greys")
```

```python
def draw_circle(x, y, r):
    yh = y - r

    while yh <= y + r:
        xh_span = int(np.sqrt(r**2 - (y - yh) ** 2))
        draw_line(x - xh_span, yh, x + xh_span, yh)
        print(f'draw line: {(x-xh_span, yh, x+xh_span, yh)}')
        yh += 1
```

```python
draw_circle(33, 44, 10)
```

```python
plt.imshow(empty_pixels, cmap='Greys')
```
